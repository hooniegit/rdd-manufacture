```python
from pyspark.sql.types import StructType, StructField, LongType, StringType, DoubleType, TimestampType

custom_schema = StructType([
    StructField("VendorID", LongType(), True),
    StructField("tpep_pickup_datetime", TimestampType(), True),
    StructField("tpep_dropoff_datetime", TimestampType(), True),
    StructField("passenger_count", LongType(), True),  # DoubleType에서 LongType으로 변경
    StructField("trip_distance", DoubleType(), True),
    StructField("RatecodeID", LongType(), True),  # DoubleType에서 LongType으로 변경
    StructField("store_and_fwd_flag", StringType(), True),
    StructField("PULocationID", LongType(), True),
    StructField("DOLocationID", LongType(), True),
    StructField("payment_type", LongType(), True),
    StructField("fare_amount", DoubleType(), True),
    StructField("extra", DoubleType(), True),
    StructField("mta_tax", DoubleType(), True),
    StructField("tip_amount", DoubleType(), True),
    StructField("tolls_amount", DoubleType(), True),
    StructField("improvement_surcharge", DoubleType(), True),
    StructField("total_amount", DoubleType(), True),
    StructField("congestion_surcharge", DoubleType(), True),
    StructField("airport_fee", DoubleType(), True),
])

df_new = spark.read.schema(custom_schema).parquet(parquet_dir)
print(df_new.count() - df_new.na.drop().count())
```

```python
df_2022_01 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-01.parquet")
df_2022_02 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-02.parquet")
df_2022_03 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-03.parquet")
df_2022_04 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-04.parquet")
df_2022_05 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-05.parquet")
df_2022_06 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-06.parquet")
df_2022_07 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-07.parquet")
df_2022_08 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-08.parquet")
df_2022_09 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-09.parquet")
df_2022_10 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-10.parquet")
df_2022_11 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-11.parquet")
df_2022_12 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-12.parquet")

df_2022 = df_2022_01.union(df_2022_02)\
                    .union(df_2022_03)\
                    .union(df_2022_04)\
                    .union(df_2022_05)\
                    .union(df_2022_06)\
                    .union(df_2022_07)\
                    .union(df_2022_08)\
                    .union(df_2022_09)\
                    .union(df_2022_10)\
                    .union(df_2022_11)\
                    .union(df_2022_12)
                    
df_2022.show()
```

```python
# 총 결측치 확인
print(df_2022.count() - df_2022.na.drop().count())
```

```python
from pyspark.sql.functions import col

# 특정 컬럼의 결측치 개수 확인
print(df_2022.filter(col("passenger_count").isNull()).count())
```

```python
# 특정 값으로 결측치 변환 / 복수의 컬럼 지정 불가능
df_2022.na.fill(0, ["passenger_count"])
```

```python
#1 : 전체 결측치 제거
df_2022.na.drop()
```

```python
# 총 중복 데이터 확인
print(df_2022.groupBy(df_2022.columns).count().filter(col("count") > 1).count())
```

```python
print(df_2022.count() - df_2022.dropDuplicates().count())
```

```python
# 총 중복 데이터 제거
df_2022.dropDuplicates()
```

```python
# 평균과 표준편차 계산
mean_value = df_2022.agg({"passenger_count": "mean"}).collect()[0][0]
stddev_value = df_2022.agg({"passenger_count": "stddev"}).collect()[0][0]

# 이상치 경계 설정 (예: 3 시그마 범위)
# lower_bound = mean_value - 3 * stddev_value
lower_bound = 0
# upper_bound = mean_value + 3 * stddev_value
upper_bound = 10

# 이상치 확인 및 출력
outliers = df_2022.filter((col("passenger_count") < lower_bound) | (col("passenger_count") > upper_bound))
outliers.show()
```

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder
from pyspark.ml import Pipeline

indexer = StringIndexer(inputCol="passenger_count", outputCol="passenger_countIndex")
encoder = OneHotEncoder(inputCol="passenger_countIndex", outputCol="passenger_count_OneHot")
pipeline = Pipeline(stages=[indexer, encoder])
model = pipeline.fit(df_2022)
encoded_df = model.transform(df_2022)

encoded_df.show()
```
