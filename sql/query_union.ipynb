```python
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession

conf = SparkConf().setAppName("query_basic")\
                  .setMaster("spark://workspace:7077")
sparkContext = SparkContext(conf=conf)
spark = SparkSession(sparkContext=sparkContext)
```

```python
parquet_dir = "../data/tlc"

df_2022_01 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-01.parquet")
df_2022_02 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-02.parquet")
df_2022_03 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-03.parquet")
df_2022_04 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-04.parquet")
df_2022_05 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-05.parquet")
df_2022_06 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-06.parquet")
df_2022_07 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-07.parquet")
df_2022_08 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-08.parquet")
df_2022_09 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-09.parquet")
df_2022_10 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-10.parquet")
df_2022_11 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-11.parquet")
df_2022_12 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2022-12.parquet")
df_2023_01 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-01.parquet")
df_2023_02 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-02.parquet")
df_2023_03 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-03.parquet")
df_2023_04 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-04.parquet")
df_2023_05 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-05.parquet")
df_2023_06 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-06.parquet")
df_2023_07 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-07.parquet")
df_2023_08 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-08.parquet")
df_2023_09 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-09.parquet")
df_2023_10 = spark.read.parquet(f"{parquet_dir}/yellow_tripdata_2023-10.parquet")

df = df_2022_01.union(df_2022_02)\
                    .union(df_2022_03)\
                    .union(df_2022_04)\
                    .union(df_2022_05)\
                    .union(df_2022_06)\
                    .union(df_2022_07)\
                    .union(df_2022_08)\
                    .union(df_2022_09)\
                    .union(df_2022_10)\
                    .union(df_2022_11)\
                    .union(df_2022_12)\
                    .union(df_2023_01)\
                    .union(df_2023_02)\
                    .union(df_2023_03)\
                    .union(df_2023_04)\
                    .union(df_2023_05)\
                    .union(df_2023_06)\
                    .union(df_2023_07)\
                    .union(df_2023_08)\
                    .union(df_2023_09)\
                    .union(df_2023_10)
```

```python
df.printSchema()
```

```python
df.createOrReplaceTempView("Taxi")
```

```python
df_1 = spark.sql(
    """
        SELECT
            t.passenger_count * 10 passengers,
            t.trip_distance * 1000 trip_kms,
            lower(t.store_and_fwd_flag) flags
        FROM
            Taxi t
    """
)
df_1.show()
```

```python
df_2 = spark.sql(
    """
        SELECT
            DISTINCT
                T.PULocationID pulocation,
                T.DOLocationID dolocation,
                T.tpep_pickup_datetime datetime
        FROM
            Taxi T
        ORDER BY
            datetime DESC
    """
)
df_2.show()
```

```python
df_2022_01.createOrReplaceTempView("df_01")
df_2022_02.createOrReplaceTempView("df_02")

df_3 = spark.sql(
    """
        SELECT
            M.VendorID,
            COUNT(*) count
        FROM
            (
                SELECT
                    *
                FROM
                    df_01 J
                UNION
                SELECT
                    *
                FROM
                    df_02 F
            ) M
        GROUP BY
            M.VendorID
    """
)
df_3.show()
```

```python
df_4 = spark.sql(
    """
        SELECT
            M.PULocationID pu,
            M.DOLocationID do,
            count(*) count
        FROM
            (
                SELECT
                    *
                FROM
                    df_01 J
                UNION
                SELECT
                    *
                FROM 
                    df_02 F
            ) M
        GROUP BY
            pu,
            do
        ORDER BY
            count DESC
    """
)
df_4.show()
```

```python
df_5 = spark.sql(
    """
        SELECT
            *
        FROM
            Taxi T
        WHERE
            (T.PULocationID >= 10 AND T.tpep_pickup_datetime >= '2023-01-01 00:00:00')
            OR
            NOT (T.DOLocationID >= 10 AND T.tpep_pickup_datetime <= '2023-01-01 00:00:00')
        ORDER BY
            2 DESC,
            5 DESC
    """
)
df_5.show()
```

```python
df_6 = spark.sql(
    """
        SELECT
            *
        FROM
            (
                SELECT
                    *
                FROM
                    df_01 J
                UNION
                SELECT
                    *
                FROM
                    df_02 F
            ) M
        WHERE
            M.tpep_pickup_datetime 
                BETWEEN 
                    '2022-01-31 00:00:00' AND '2022-02-01 00:00:00'
        ORDER BY
            M.tpep_pickup_datetime DESC
    """
)
df_6.show()
```

```python
df_7 = spark.sql(
    """
        SELECT
            *
        FROM
            Taxi T
        WHERE
            T.tpep_pickup_datetime
                IN
                    (
                        SELECT 
                            J.tpep_pickup_datetime
                        FROM
                            df_01 J
                        UNION
                        SELECT
                            F.tpep_pickup_datetime
                        FROM
                            df_02 F
                    )
        ORDER BY
            T.tpep_pickup_datetime DESC
    """
)
df_7.show()
```

```python
df_08 = spark.sql(
    """
        SELECT
            DISTINCT
                *
        FROM
            Taxi T
        WHERE
            T.tpep_pickup_datetime
                LIKE
                    "2023-08%"
        ORDER BY
            T.tpep_pickup_datetime DESC
    """
)
df_08.show()
```

```python
df_09 = spark.sql(
    """
        SELECT
            MONTH(tpep_pickup_datetime) pickup_month,
            count(*) count
        FROM
            (
                SELECT
                    *
                FROM
                    Taxi T
                WHERE
                    T.tpep_pickup_datetime
                        LIKE
                            "2023-__-01%"
            ) F
        GROUP BY
            pickup_month
        ORDER BY 
            count DESC  
    """
)
df_09.show()
```

```python
df_10 = spark.sql(
    """
        SELECT
            MONTH(tpep_pickup_datetime) month,
            MAX(passenger_count) max_passenger,
            count(*) count
        FROM
            Taxi T
        GROUP BY
            month
        HAVING
            count >= 1000
        ORDER BY
            count DESC
    """
)
df_10.show()
```
